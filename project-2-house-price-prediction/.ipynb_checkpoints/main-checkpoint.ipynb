## 1. Очистка, предварительная обработка данных

````python
import numpy as np
import pandas as pd
````

````python
train_data = pd.read_csv("train.csv")
test_data = pd.read_csv("test.csv")
````

````python
train_data
````

````python
test_data
````

````python
train_data.info()
````

````python
test_data.info()
````

Колонка `Id` не несет никакой полезной информации. Много значений `NaN` в некоторых колонках. Существуют значения, отличные от представленных в `data_description.txt`. К примеру, в `train_data["MSSubClass"]` должно быть значение `C` вместо `C (all)`

````python
for column in train_data.columns:
    print(column, train_data[column].unique())
````

````python
for column in test_data.columns:
    print(column, test_data[column].unique())
````

Проверив уникальные значения каждого столбца для `train_data`, было замечено, что некоторые значения не совсем совпадают со значениями из `data_description.txt`. К примеру, в `train_data["Exterior2nd"]` значение `Wd Shing` должно быть `WdShing`, `CmentBd` должно быть `CemntBd`, `Brk Cmn` должно быть `BrkComm`. Однако, проверив уникальные значения и для `test_data`, было замечено, что эти ошибки одинаковы для всех данных и в `train_data` и в `test_data`, поэтому, переименование не было произведено. 

Значения `NaN` были заменены на моду для категорийных данных или `NA`, если это указано в `data_description.txt`. Например для колонки `Alley` есть специальная метка `NA = No alley access`. Для цифровых данных, было взято среднее значение

````python
train_data["LotFrontage"].fillna(float(int(train_data["LotFrontage"].mean())), inplace=True)
test_data["LotFrontage"].fillna(float(int(test_data["LotFrontage"].mean())), inplace=True)

train_data["Alley"].fillna("NA", inplace=True)
test_data["Alley"].fillna("NA", inplace=True)

train_data["MasVnrType"].fillna("None", inplace=True)
test_data["MasVnrType"].fillna("None", inplace=True)

train_data["MasVnrArea"].fillna(float(int(train_data["MasVnrArea"].mean())), inplace=True)
test_data["MasVnrArea"].fillna(float(int(test_data["MasVnrArea"].mean())), inplace=True)

train_data["BsmtQual"].fillna("NA", inplace=True)
test_data["BsmtQual"].fillna("NA", inplace=True)

train_data["BsmtCond"].fillna("NA", inplace=True)
test_data["BsmtCond"].fillna("NA", inplace=True)

train_data["BsmtExposure"].fillna("NA", inplace=True)
test_data["BsmtExposure"].fillna("NA", inplace=True)

train_data["BsmtFinType1"].fillna("NA", inplace=True)
test_data["BsmtFinType1"].fillna("NA", inplace=True)

train_data["BsmtFinType2"].fillna("NA", inplace=True)
test_data["BsmtFinType2"].fillna("NA", inplace=True)

train_data["Electrical"].fillna(train_data["Electrical"].mode()[0], inplace=True)

train_data["FireplaceQu"].fillna(train_data["FireplaceQu"].mode()[0], inplace=True)
test_data["FireplaceQu"].fillna(test_data["FireplaceQu"].mode()[0], inplace=True)

train_data["GarageType"].fillna("NA", inplace=True)
test_data["GarageType"].fillna("NA", inplace=True)

train_data["GarageYrBlt"].fillna(float(int(train_data["GarageYrBlt"].mean())), inplace=True)
test_data["GarageYrBlt"].fillna(float(int(test_data["GarageYrBlt"].mean())), inplace=True)

train_data["GarageFinish"].fillna("NA", inplace=True)
test_data["GarageFinish"].fillna("NA", inplace=True)

train_data["GarageQual"].fillna("NA", inplace=True)
test_data["GarageQual"].fillna("NA", inplace=True)

train_data["GarageCond"].fillna("NA", inplace=True)
test_data["GarageCond"].fillna("NA", inplace=True)

train_data["PoolQC"].fillna("NA", inplace=True)
test_data["PoolQC"].fillna("NA", inplace=True)

train_data["Fence"].fillna("NA", inplace=True)
test_data["Fence"].fillna("NA", inplace=True)

train_data["MiscFeature"].fillna("NA", inplace=True)
test_data["MiscFeature"].fillna("NA", inplace=True)
````

````python
train_data.isna().values.any()
````

````python
test_data.isna().values.any()
````

````python
categorical_cols = []
quantitative_cols = []

for column, dtype in zip(train_data.dtypes.index, train_data.dtypes):
    if dtype == "object":
        categorical_cols.append(column)
    else:
        quantitative_cols.append(column)
        
if "SalePrice" in categorical_cols:
    categorical_cols.remove("SalePrice")
if "SalePrice" in quantitative_cols:
    quantitative_cols.remove("SalePrice")
    
print(f"Categorical columns:\n{categorical_cols}\n")
print(f"Quatitative columns:\n{quantitative_cols}")
````

### Перевод категорийных данных в числовой тип

````python
from sklearn.preprocessing import LabelEncoder
````

````python
#encoder = LabelEncoder()

for column in categorical_cols:
    encoder = LabelEncoder()
    train_data[column] = encoder.fit_transform(train_data[column].values)
    test_data[column] = encoder.fit_transform(test_data[column].values)
````

````python
train_data
````

````python
test_data
````

### Выбор топ 10 параметров в отношении `SalePrice`

````python
from sklearn.feature_selection import mutual_info_regression
````

````python
top_10 = mutual_info_regression(train_data[train_data.columns.drop("SalePrice")], train_data["SalePrice"])
top_10
````

````python
top_10 = pd.Series(top_10)
top_10.index = train_data.columns.drop("SalePrice")
top_10 = top_10.sort_values(ascending=False)[:10].index
````

````python
top_10
````

### Стандартизация данных

````python
from sklearn.preprocessing import StandardScaler
````

````python
scaled_train_data = train_data[top_10].copy()
scaled_test_data = test_data[top_10].copy()
scaled_train_data["SalePrice"] = train_data["SalePrice"]
scaled_test_data["SalePrice"] = test_data["SalePrice"]
````

````python
sc_x = StandardScaler()
sc_y = StandardScaler()

sc_x.fit(scaled_train_data[top_10].values)
scaled_train_data[top_10] = sc_x.transform(scaled_train_data[top_10].values)
    
sc_y.fit(scaled_train_data["SalePrice"].values.reshape(-1, 1))
scaled_train_data["SalePrice"] = sc_y.transform(scaled_train_data["SalePrice"].values.reshape(-1, 1))
````

````python
scaled_train_data
````

````python
scaled_test_data[top_10] = sc_x.transform(scaled_test_data[top_10].values)
````

````python
scaled_test_data
````

## 2. Обучение и проверка моделей

````python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
````

````python
import sklearn.metrics as metrics
from sklearn.linear_model import Lasso
from sklearn.linear_model import ElasticNet
from sklearn.neural_network import MLPRegressor
````

````python
scaled_lasso_reg = Lasso()
scaled_lasso_reg.fit(scaled_train_data[top_10].values, scaled_train_data["SalePrice"].values.reshape(-1,1))
scaled_lasso_pred = scaled_lasso_reg.predict(scaled_test_data[top_10].values)
scaled_lasso_pred = sc_y.inverse_transform(scaled_lasso_pred.reshape(-1,1))
````

````python
lasso_reg = Lasso()
lasso_reg.fit(train_data[top_10].values, train_data["SalePrice"].values.reshape(-1,1))
lasso_pred = lasso_reg.predict(test_data[top_10].values)
````

````python
scaled_EN_reg = ElasticNet()
scaled_EN_reg.fit(scaled_train_data[top_10].values, scaled_train_data["SalePrice"].values.reshape(-1,1))
scaled_EN_pred = scaled_EN_reg.predict(scaled_test_data[top_10].values)
scaled_EN_pred = sc_y.inverse_transform(scaled_EN_pred.reshape(-1,1))
````

````python
EN_reg = ElasticNet()
EN_reg.fit(train_data[top_10].values, train_data["SalePrice"].values.reshape(-1,1))
EN_pred = EN_reg.predict(test_data[top_10].values)
````

````python
scaled_MLP_reg = MLPRegressor(max_iter=4500)
scaled_MLP_reg.fit(scaled_train_data[top_10].values, scaled_train_data["SalePrice"].values)
scaled_MLP_pred = scaled_MLP_reg.predict(scaled_test_data[top_10].values)
scaled_MLP_pred = sc_y.inverse_transform(scaled_MLP_pred.reshape(-1,1))
````

````python
MLP_reg = MLPRegressor(max_iter=5000)
MLP_reg.fit(train_data[top_10].values, train_data["SalePrice"].values)
MLP_pred = MLP_reg.predict(test_data[top_10].values)
````

````python
plt.figure(figsize=(9,7))
plt.scatter(test_data["SalePrice"].values, scaled_lasso_pred, color="blue", label="Lasso (scaled data)", edgecolor="white");
plt.scatter(test_data["SalePrice"].values, scaled_EN_pred, color="darkorange", label="ElasticNet (scaled data)", edgecolor="white");
plt.scatter(test_data["SalePrice"].values, scaled_MLP_pred, color="green", label="MPL (scaled data)", edgecolor="white");
plt.xlabel("True house price");
plt.ylabel("Predicted house price");
plt.title("True vs Predicted");
plt.legend();
````

````python
plt.figure(figsize=(9,7))
plt.scatter(test_data["SalePrice"].values, lasso_pred, color="blue", label="Lasso", edgecolor="white");
plt.scatter(test_data["SalePrice"].values, EN_pred, color="darkorange", label="ElasticNet", edgecolor="white");
plt.scatter(test_data["SalePrice"].values, MLP_pred, color="green", label="MPL", edgecolor="white");
plt.xlabel("True house price");
plt.ylabel("Predicted house price");
plt.title("True vs Predicted");
plt.legend();
````

````python
plt.figure(figsize=(9,7))
plt.scatter(test_data["SalePrice"].values, lasso_pred, color="blue", label="Lasso", edgecolor="white");
plt.scatter(test_data["SalePrice"].values, EN_pred, color="darkorange", label="ElasticNet", edgecolor="white");
plt.scatter(test_data["SalePrice"].values, scaled_MLP_pred, color="green", label="MPL (scaled data)", edgecolor="white");
plt.xlabel("True house price");
plt.ylabel("Predicted house price");
plt.title("True vs Predicted");
plt.legend();
````

- Стандартизированные данные ухудшают предсказания для моделей `Lasso` и `ElasticNet`
- Стандартизированные данные улучшают предсказания для модели `MLP`

## 3. Сравнение моделей на основе метрик регрессии

````python
print("Lasso metrics:")
print("MSE =", metrics.mean_squared_error(test_data["SalePrice"].values, lasso_pred))
print("RMSE =", metrics.mean_squared_error(test_data["SalePrice"].values, lasso_pred, squared=False))
print("MAE =", metrics.mean_absolute_error(test_data["SalePrice"].values, lasso_pred))
print("R2 =", metrics.r2_score(test_data["SalePrice"].values, lasso_pred))
print("R2_adjusted =", 1 - (1 - metrics.r2_score(test_data["SalePrice"].values, lasso_pred)) * (len(lasso_pred) - 1) / (len(lasso_pred) - 10 - 1))
````

````python
print("ElasticNet metrics:")
print("MSE =", metrics.mean_squared_error(test_data["SalePrice"].values, EN_pred))
print("RMSE =", metrics.mean_squared_error(test_data["SalePrice"].values, EN_pred, squared=False))
print("MAE =", metrics.mean_absolute_error(test_data["SalePrice"].values, EN_pred))
print("R2 =", metrics.r2_score(test_data["SalePrice"].values, EN_pred))
print("R2_adjusted =", 1 - (1 - metrics.r2_score(test_data["SalePrice"].values, EN_pred)) * (len(EN_pred) - 1) / (len(EN_pred) - 10 - 1))
````

````python
print("MLP metrics:")
print("MSE =", metrics.mean_squared_error(test_data["SalePrice"].values, scaled_MLP_pred))
print("RMSE =", metrics.mean_squared_error(test_data["SalePrice"].values, scaled_MLP_pred, squared=False))
print("MAE =", metrics.mean_absolute_error(test_data["SalePrice"].values, scaled_MLP_pred))
print("R2 =", metrics.r2_score(test_data["SalePrice"].values, scaled_MLP_pred))
print("R2_adjusted =", 1 - (1 - metrics.r2_score(test_data["SalePrice"].values, scaled_MLP_pred)) * (len(scaled_MLP_pred) - 1) / (len(scaled_MLP_pred) - 10 - 1))
````
