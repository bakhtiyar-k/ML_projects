### Перевод категорийных данных в числовой тип

````python
from sklearn.preprocessing import LabelEncoder
````

````python
for column in categorical_cols:
    encoder = LabelEncoder()
    train_data[column] = encoder.fit_transform(train_data[column].values)
    test_data[column] = encoder.fit_transform(test_data[column].values)
````

````python
train_data
````

````python
test_data
````

### Выбор параметров в отношении `SalePrice`

````python
sns.heatmap(train_data.corr(), annot=True, linewidth=0.5);
````

- Так как у нас мало параметров и они не сильно коррелируют между собой, возьмем все для обучения модели

### Стандартизация данных

````python
from sklearn.preprocessing import StandardScaler
````

````python
scaled_train_data = train_data.copy()
scaled_test_data = test_data.copy()
````

````python
params = scaled_train_data[scaled_train_data.columns[:-1]].columns
params
````

````python
sc_x = StandardScaler()
sc_y = StandardScaler()

sc_x.fit(scaled_train_data[params].values)
scaled_train_data[params] = sc_x.transform(scaled_train_data[params].values)

sc_y.fit(scaled_train_data["charges"].values.reshape(-1, 1))
scaled_train_data["charges"] = sc_y.transform(scaled_train_data["charges"].values.reshape(-1, 1))
````

````python
scaled_train_data
````

````python
scaled_test_data[params] = sc_x.transform(scaled_test_data[params].values)
````

````python
scaled_test_data
````

## 2. Обучение и проверка моделей

````python
from sklearn.linear_model import Ridge
from sklearn.ensemble import AdaBoostRegressor
from sklearn.cross_decomposition import PLSRegression
````

````python
scaled_ridge_reg = Ridge()
scaled_ridge_reg.fit(scaled_train_data[params].values, scaled_train_data["charges"].values.reshape(-1, 1))
scaled_ridge_pred = scaled_ridge_reg.predict(scaled_test_data[params].values)
scaled_ridge_pred = sc_y.inverse_transform(scaled_ridge_pred.reshape(-1,1))
````

````python
ridge_reg = Ridge()
ridge_reg.fit(train_data[params].values, train_data["charges"].values.reshape(-1,1))
ridge_pred = ridge_reg.predict(test_data[params].values)
````

````python
scaled_ABR_reg = AdaBoostRegressor()
scaled_ABR_reg.fit(scaled_train_data[params].values, scaled_train_data["charges"].values)
scaled_ABR_pred = scaled_ABR_reg.predict(scaled_test_data[params].values)
scaled_ABR_pred = sc_y.inverse_transform(scaled_ABR_pred.reshape(-1,1))
````

````python
ABR_reg = AdaBoostRegressor()
ABR_reg.fit(train_data[params].values, train_data["charges"].values)
ABR_pred = ABR_reg.predict(test_data[params].values)
````

````python
scaled_PLS_reg = PLSRegression()
scaled_PLS_reg.fit(scaled_train_data[params].values, scaled_train_data["charges"].values)
scaled_PLS_pred = scaled_PLS_reg.predict(scaled_test_data[params].values)
scaled_PLS_pred = sc_y.inverse_transform(scaled_PLS_pred.reshape(-1,1))
````

````python
PLS_reg = PLSRegression()
PLS_reg.fit(train_data[params].values, train_data["charges"].values)
PLS_pred = PLS_reg.predict(test_data[params].values)
````

````python
plt.figure(figsize=(9,7))
plt.scatter(test_data["charges"].values, scaled_ridge_pred, color="blue", label="Ridge (scaled data)", edgecolor="white");
plt.scatter(test_data["charges"].values, scaled_ABR_pred, color="darkorange", label="AdaBoostRegressor (scaled data)", edgecolor="white");
plt.scatter(test_data["charges"].values, scaled_PLS_pred, color="green", label="PSLRegressor (scaled data)", edgecolor="white");
plt.xlabel("True charges");
plt.ylabel("Predicted charges");
plt.title("True vs Predicted");
plt.legend();
````

````python
plt.figure(figsize=(9,7))
plt.scatter(test_data["charges"].values, ridge_pred, color="blue", label="Ridge", edgecolor="white");
plt.scatter(test_data["charges"].values, ABR_pred, color="darkorange", label="AdaBoostRegression", edgecolor="white");
plt.scatter(test_data["charges"].values, PLS_pred, color="green", label="PLSRegression", edgecolor="white");
plt.xlabel("True charges");
plt.ylabel("Predicted charges");
plt.title("True vs Predicted");
plt.legend();
````

- Стандартизация не влияет на предсказания моделей
- `Ridge` в точности повторяет предсказания `PLSRegression`

## 3. Сравнение моделей на основе метрик регрессии

````python
import sklearn.metrics as metrics
````

````python
print("Ridge metrics:")
print("MSE =", metrics.mean_squared_error(test_data["charges"].values, ridge_pred))
print("RMSE =", metrics.mean_squared_error(test_data["charges"].values, ridge_pred, squared=False))
print("MAE =", metrics.mean_absolute_error(test_data["charges"].values, ridge_pred))
print("R2 =", metrics.r2_score(test_data["charges"].values, ridge_pred))
print("R2_adjusted =", 1 - (1 - metrics.r2_score(test_data["charges"].values, ridge_pred)) * (len(ridge_pred) - 1) / (len(ridge_pred) - 6 - 1))
````

````python
print("AdaBoostRegression metrics:")
print("MSE =", metrics.mean_squared_error(test_data["charges"].values, ABR_pred))
print("RMSE =", metrics.mean_squared_error(test_data["charges"].values, ABR_pred, squared=False))
print("MAE =", metrics.mean_absolute_error(test_data["charges"].values, ABR_pred))
print("R2 =", metrics.r2_score(test_data["charges"].values, ABR_pred))
print("R2_adjusted =", 1 - (1 - metrics.r2_score(test_data["charges"].values, ABR_pred)) * (len(ABR_pred) - 1) / (len(ABR_pred) - 6 - 1))
````

````python
print("PLSRegressor metrics:")
print("MSE =", metrics.mean_squared_error(test_data["charges"].values, PLS_pred))
print("RMSE =", metrics.mean_squared_error(test_data["charges"].values, PLS_pred, squared=False))
print("MAE =", metrics.mean_absolute_error(test_data["charges"].values, PLS_pred))
print("R2 =", metrics.r2_score(test_data["charges"].values, PLS_pred))
print("R2_adjusted =", 1 - (1 - metrics.r2_score(test_data["charges"].values, PLS_pred)) * (len(PLS_pred) - 1) / (len(PLS_pred) - 6 - 1))
````

- `AdaBoostRegressor` предсказывает точнее по всем метрикам
