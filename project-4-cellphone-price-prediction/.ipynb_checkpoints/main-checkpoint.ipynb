````python
important = ["ram", "battery_power", "px_height", "px_width"]
````

### Стандартизация данных

````python
from sklearn.preprocessing import StandardScaler
````

````python
scaled_train_data = train_data.copy()
scaled_test_data = test_data.copy()
````

````python
sc_x = StandardScaler()
sc_x.fit(scaled_train_data[important].values)
scaled_train_data[important] = sc_x.transform(scaled_train_data[important].values)
scaled_test_data[important] = sc_x.transform(scaled_test_data[important].values)
````

````python
scaled_train_data[important]
````

````python
scaled_test_data[important]
````

## 2. Обучение и проверка моделей

````python
from sklearn.linear_model import RidgeClassifier
from sklearn.linear_model import SGDClassifier
from sklearn.neighbors import NearestCentroid
````

````python
ridge_clf = RidgeClassifier()
ridge_clf.fit(scaled_train_data[important].values, scaled_train_data["price_range"].values)
ridge_pred = ridge_clf.predict(scaled_test_data[important].values)
````

````python
SGD_clf = SGDClassifier()
SGD_clf.fit(scaled_train_data[important].values, scaled_train_data["price_range"])
SGD_pred = SGD_clf.predict(scaled_test_data[important].values)
````

````python
NC_clf = NearestCentroid()
NC_clf.fit(scaled_train_data[important].values, scaled_train_data["price_range"].values)
NC_pred = NC_clf.predict(scaled_test_data[important].values)
````

## 3. Сравнение моделей на основе метрик классификации: F-score, Precision, Recall

````python
from sklearn.metrics import precision_score, recall_score, f1_score
````

````python
precision_ridge = precision_score(test_data["price_range"].values, ridge_pred, average="macro")
precision_SGD = precision_score(test_data["price_range"].values, SGD_pred, average="macro")
precision_NC = precision_score(test_data["price_range"].values, NC_pred, average="macro")

recall_ridge = recall_score(test_data["price_range"].values, ridge_pred, average="macro")
recall_SGD = recall_score(test_data["price_range"].values, SGD_pred, average="macro")
recall_NC = recall_score(test_data["price_range"].values, NC_pred, average="macro")

f1_ridge = f1_score(test_data["price_range"].values, ridge_pred, average="macro")
f1_SGD = f1_score(test_data["price_range"].values, SGD_pred, average="macro")
f1_NC = f1_score(test_data["price_range"].values, NC_pred, average="macro")

print(f"RidgeClassifier precision : {precision_ridge:.2f}")
print(f"SGDClassifier precision: {precision_SGD:.2f}")
print(f"NearestCentroid precision: {precision_NC:.2f}\n")

print(f"RidgeClassifier recall: {recall_ridge:.2f}")
print(f"SGDClassifier recall: {recall_SGD:.2f}")
print(f"NearestCentroid recall: {recall_NC:.2f}\n")

print(f"RidgeClassifier F1 Score: {f1_ridge:.2f}")
print(f"SGDClassifier F1 Score: {f1_SGD:.2f}")
print(f"NearestCentroid F1 Score: {f1_NC:.2f}")
````

- `RidgeClassifier` наихудший по всем метрикам
- `SGDClassifier` и `NearestCentroid` по оценкам метрик почти одинаковы
