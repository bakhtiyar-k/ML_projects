## 1. Очистка, предварительная обработка данных, и Exploratory Data Analysis

````python
import numpy as np
import pandas as pd
````

````python
train_data = pd.read_csv("train_data.csv")
test_data = pd.read_csv("test_data.csv")
````

````python
train_data["Annual Income"].fillna(float(int(train_data["Annual Income"].mean())), inplace=True)
train_data["Years in current job"].fillna(train_data["Years in current job"].mode()[0], inplace=True)
train_data["Months since last delinquent"].fillna(float(int(train_data["Months since last delinquent"].mean())), inplace=True)
train_data["Bankruptcies"].fillna(float(int(train_data["Bankruptcies"].mean())), inplace=True)
train_data["Credit Score"].fillna(float(int(train_data["Credit Score"].mean())), inplace=True)
````

````python
train_data.isnull().sum()
````

````python
test_data.info()
````

````python
test_data["Annual Income"].fillna(float(int(test_data["Annual Income"].mean())), inplace=True)
test_data["Years in current job"].fillna(test_data["Years in current job"].mode()[0], inplace=True)
test_data["Months since last delinquent"].fillna(float(int(test_data["Months since last delinquent"].mean())), inplace=True)
test_data["Bankruptcies"].fillna(float(int(test_data["Bankruptcies"].mean())), inplace=True)
test_data["Credit Score"].fillna(float(int(test_data["Credit Score"].mean())), inplace=True)
````

````python
test_data.isnull().sum()
````

````python
categorical_cols = train_data.select_dtypes(include=["object"]).columns.to_list()
numerical_cols = train_data.select_dtypes(exclude=["object"]).columns.to_list()

print("Categorical columns:", categorical_cols, "\n")
print("Numerical columns:", numerical_cols)
````

### Перевод категорийных данных в численный тип

````python
from sklearn.preprocessing import LabelEncoder
````

````python
encoded_train_data = train_data.copy()
encoded_test_data = test_data.copy()
````

````python
for column in categorical_cols:
    encoder = LabelEncoder()
    encoded_train_data[column] = encoder.fit_transform(encoded_train_data[column])
    encoded_test_data[column] = encoder.fit_transform(encoded_test_data[column])
````

````python
important = ["Term", "Current Loan Amount", "Credit Score"]
````

### Стандартизация данных

````python
from sklearn.preprocessing import StandardScaler
````

````python
scaled_train_data = encoded_train_data.copy()
scaled_test_data = encoded_test_data.copy()
````

Нет надобности стандартизировать колонку `Term`, так как его уникальные значения 0 и 1

````python
sc_x = StandardScaler()
sc_x.fit(scaled_train_data[important].drop("Term", axis=1).values, scaled_train_data["Credit Default"].values)
scaled_train_data[["Current Loan Amount", "Credit Score"]] = sc_x.transform(scaled_train_data[important].drop("Term", axis=1).values)
scaled_test_data[["Current Loan Amount", "Credit Score"]] = sc_x.transform(scaled_test_data[important].drop("Term", axis=1).values)
````

````python
scaled_train_data[important]
````

````python
scaled_test_data[important]
````

## 2. Обучение и проверка моделей

````python
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.neural_network import MLPClassifier
````

````python
gaussian_clf = GaussianProcessClassifier()
gaussian_clf.fit(scaled_train_data[important].values, scaled_train_data["Credit Default"].values)
gaussian_pred = gaussian_clf.predict(scaled_test_data[important].values)
````

````python
grad_boost_clf = GradientBoostingClassifier()
grad_boost_clf.fit(scaled_train_data[important].values, scaled_train_data["Credit Default"].values)
grad_boost_pred = grad_boost_clf.predict(scaled_test_data[important].values)
````

````python
MLP_clf = MLPClassifier()
MLP_clf.fit(scaled_train_data[important].values, scaled_train_data["Credit Default"].values)
MLP_pred = MLP_clf.predict(scaled_test_data[important].values)
````

## 3. Сравнение моделей на основе метрик классификации: F-score, Precision, Recall

````python
from sklearn.metrics import precision_score, recall_score, f1_score
````

````python
precision_gaussian = precision_score(test_data["Credit Default"].values, gaussian_pred)
precision_grad_boost = precision_score(test_data["Credit Default"].values, grad_boost_pred)
precision_MLP = precision_score(test_data["Credit Default"].values, MLP_pred)

recall_gaussian = recall_score(test_data["Credit Default"].values, gaussian_pred)
recall_grad_boost = recall_score(test_data["Credit Default"].values, grad_boost_pred)
recall_MLP = recall_score(test_data["Credit Default"].values, MLP_pred)

f1_gaussian = f1_score(test_data["Credit Default"].values, gaussian_pred)
f1_grad_boost = f1_score(test_data["Credit Default"].values, grad_boost_pred)
f1_MLP = f1_score(test_data["Credit Default"].values, MLP_pred)

print(f"GaussianProcessClassifier precision : {precision_gaussian:.2f}")
print(f"GradientBoostingClassifier precision: {precision_grad_boost:.2f}")
print(f"MLPClassifier precision: {precision_MLP:.2f}\n")

print(f"GaussianProcessClassifier recall: {recall_gaussian:.2f}")
print(f"GradientBoostingClassifier recall: {recall_grad_boost:.2f}")
print(f"MLPClassifier recall: {recall_MLP:.2f}\n")

print(f"GaussianProcessClassifier F1 Score: {f1_gaussian:.2f}")
print(f"GradientBoostingClassifier F1 Score: {f1_grad_boost:.2f}")
print(f"MLPClassifier F1 Score: {f1_MLP:.2f}")
````

- Все три модели предсказывают примерно одинаково
